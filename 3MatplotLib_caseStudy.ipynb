#### [8] Plot Month_Job
Plot the Bar graph between Month vs Job Openings

Print the month name and the number of job posting as integer value. Order of months doesn't matter.

```python
#Output Format : month_name1 job_posting1
import csv
import collections
with open('/content/drive/MyDrive/Colab Notebooks/10amazon_jobs_dataset.csv') as file_obj:
    file_data = csv.DictReader(file_obj, skipinitialspace = True)
    dct = {}
    for row in file_data:
        date = row['Posting_date'].split()
        key = date[0]
        if key in dct:
            dct[key] += 1
        else:
            dct[key] = 1
    month_jobs = collections.OrderedDict(sorted(dct.items())) 
for month in month_jobs:
    print(month, month_jobs[month], end=', ')
```

```python
import matplotlib.pyplot as plt
plt.bar( list(month_jobs.keys()), list(month_jobs.values()))
plt.xlabel("Month")
plt.ylabel("Jobs")
plt.title('Month vs Job_Openings')
plt.xticks(rotation = 45)
plt.show()    
```

#### [9] Plot city_job
Plot the Pie chart between Indian cities vs No. of jobs opening.

Print the Indian cities and %age of Job distribution in India upto 2 decimal places. Note: %age of Job distribution should be in descending order.

```python
#Output Format :  city1 percentage1
import csv
import collections
with open('/content/drive/MyDrive/Colab Notebooks/10amazon_jobs_dataset.csv') as file_obj:
    file_data = csv.DictReader(file_obj, skipinitialspace = True)
    dct = {}
    for row in file_data:
        location = row['location'].split(',')
        country = location[0]
        if country=='IN' :
            city = location[2]
            if city in dct:
                dct[city] += 1
            else:
                dct[city] = 1
    ord_dct = collections.OrderedDict(sorted(dct.items()))                                  
    print(ord_dct)           
```

```python
import numpy as np
city = list(ord_dct.keys())
job = np.array(list(ord_dct.values()))
job_pct = np.divide(job, job.sum()) * 100
for i in range(len(job)):
    print(city[i],format(job_pct[i], '.2f'), end=', ')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv

file_obj = open("/content/drive/MyDrive/Colab Notebooks/10amazon_jobs_dataset.csv")
df = pd.read_csv(file_obj)
file_obj.close()

location = df['location'].str.split(', ', expand = True)
bool_arr = location[0] == 'IN'
india_city = location[bool_arr][2]
freq = india_city.value_counts()
x = freq.index
plt.axis('equal')
plt.xticks(rotation = 40)
plt.pie(freq, labels = x, autopct = '%.2f%%')
plt.show()

perc = np.true_divide(freq, freq.sum(axis = 0))*100
for i in range(len(x)):
    print(x[i], format(perc[i], '0.2f'), end = ' ')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv

file_obj = open('/content/drive/MyDrive/Colab Notebooks/10amazon_jobs_dataset.csv')
df = pd.read_csv(file_obj)
file_obj.close()

year = df['Posting_date'].str.split(', ', expand = True)[1]
skills = df['BASIC QUALIFICATIONS']
arr = np.array(list(zip(year, skills)))

for i in range(len(arr)):
    if 'Java' in arr[i][1] or 'java' in arr[i][1]:
        arr[i][1] = 1
    else :
        arr[i][1] = 0

skill_year = arr[arr[:, 1] == '1']
year_With_java_skill = skill_year[:, 0]

year = np.unique(year_With_java_skill)
job = []
for i in year:
    job.append(len(year_With_java_skill[i == year_With_java_skill]))

plt.scatter(year, job, edgecolors = 'black', s=100)
plt.grid()
plt.show()

for i in range(len(year)):
    print(year[i], job[i], end=', ')
```
